# Functions

## Helper Functions


```python
def calculate_similarity(letters, labels):
    similarity_matrix = []
    
    for i, letter1 in enumerate(letters):
        inproducts = [] 
        
        for j, letter2 in enumerate(letters):
            inproduct = np.dot(letter1, letter2)
            inproducts.append(inproduct)
            
        similarity_matrix.append(inproducts)
        
    return similarity_matrix
    

def evaluate(network, letter_matrix, labels, classification_labels, actual_index):
    matrix = np.matmul(network, letter_matrix)
    predicted = np.argmax(matrix)
    
    if np.all(matrix == matrix[0]):
        return labels[actual_index], '', matrix
    return labels[actual_index], classification_labels[predicted], matrix


def normalize(x):
    len_x = np.sqrt(np.dot(x, x))
    return (1/len_x) * x


def reshape_letter_matrix(letter, size=5):
    return np.reshape(letter, (size, size))


def flatten_matrix(matrix):
    return np.matrix.flatten(matrix)


def get_threshold(matrix):
    return int(np.min(np.array(matrix).diagonal()))


def normalize_matrix(matrix):
    normalize = np.linalg.norm(matrix)
    normalized_matrix = matrix/normalize
    return normalized_matrix
```

## Create Plots - Functions


```python
def display_bar_chart(input, labels, draw_threshold=False):
    width = np.arange(len(labels))
    bar_width = 0.25
    opacity = 0.4
    ax = plt.subplot()
    threshold=None

    if draw_threshold: 
        threshold = get_threshold(input)

    for i, matrix in enumerate(input):
        ax.bar(width  + i/4, matrix, width=bar_width,
            alpha=opacity,
            label=labels[i])
    
    if threshold: ax.axhline(threshold, color="gray")
    plt.xticks(width, labels)
    ax.legend()
    plt.show()
    


def create_similarity_matrix(matrix, labels, columns, title="Similarity Matrix", figure_size=None):
    df = pd.DataFrame(matrix, index = labels,
                      columns = columns)
    plt.figure(figsize=figure_size)
    sn.heatmap(df, annot=True, fmt='g', cmap='Greens')
    plt.title(title)
    plt.xlabel('predicted')
    plt.ylabel('actual')
    plt.show()

def create_confusion_matrix(actual_array, predicted_array, labels):
    matrix = confusion_matrix(actual_array, predicted_array, labels=labels)
    df = pd.DataFrame(matrix, index = labels,
                      columns = labels)
    plt.figure()
    sn.heatmap(df, annot=True, fmt='g', cmap='Greens')
    plt.title('Confusion Matrix')
    plt.xlabel('predicted')
    plt.ylabel('actual')
    plt.show()


def create_letter_plot(letter):
    matrix = reshape_letter_matrix(letter)
    fig, ax = plt.subplots()
    cax = ax.matshow(matrix, vmin=0, vmax=0.99, cmap='YlGn')
    # fig.colorbar(cax)
    for (i, j), z in np.ndenumerate(matrix):
        ax.text(j, i, '{:0.2f}'.format(z), ha='center', va='center')
    plt.show()
```

## Create variations of the basic characters

```python
def shift_matrix(letter): 
    matrix = np.copy(letter)
    x = reshape_letter_matrix(matrix)
    rolled_matrix = np.roll(x, 1, axis=1)
    return flatten_matrix(rolled_matrix)


def blur_matrix(letter):
    matrix = np.copy(letter)
    blur_matrix = create_blur(25)
    blur_result = np.matmul(blur_matrix, matrix)
    # double_blur = np.matmul(blur_matrix, blur_result)
    return blur_result


def remove_first_last_row(letter):
    matrix = np.copy(letter)
    length = len(matrix)
    length_row = 5
    zero_matrix = np.zeros(length_row)
    matrix[length-length_row:] = zero_matrix # replace last x and first x lines of vector 
    matrix[:length_row] = zero_matrix
    return matrix

def remove_last_row(letter, amount):
    matrix = np.copy(letter)
    length = len(matrix)
    length_row =amount * 5
    zero_matrix = np.zeros(length_row)
    matrix[length-length_row:] = zero_matrix # replace last x lines of vector 
    return matrix

def add_brightness(letter):
    matrix = np.copy(letter)
    result = [number - int((number/2)*100)/100 for number in matrix]
    return result

def combine_matrices(a, b, c):
    if c is not None:
        x = np.maximum.reduce([a, b, c])
    else: x = np.maximum.reduce([a, b])

    return x
```

## Pre-Processing Functions

```python
# Blur matrix (copy of function from W2.1 Blurring explanation.ipynb)
def create_blur(size):
    blur_M = np.zeros((size,size))
    if size > 1:
        for i in range(0, blur_M.shape[1]): 
            if i == 0:                             # EDGE CASE : UPPER LEFT  
                blur_M[i,i] = 0.8
                blur_M[1,i] = 0.2
            elif i == (blur_M.shape[1] - 1):       # EDGE CASE : LOWER RIGHT
                blur_M[i, i] = 0.8
                blur_M[i-1, i] = 0.2
            else:                                  # MAIN CASE : ENTRIES
                blur_M[i,i] = 0.6                  # MIDDLE VALUE
                blur_M[i-1,i] = 0.2                # LEFT VALUE
                blur_M[i+1,i] = 0.2                # RIGHT VALUE

    return blur_M


def negative_correlation(matrix):
    m_minus = np.copy(matrix)
    for i in range(len(m_minus)):
        m_minus[i] -= 0.26
    
    return m_minus
```
